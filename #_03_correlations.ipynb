```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\Code\PipPandas\outputs\public_companies.csv')
```

```python
# looking for any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing)) 
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
correlation_matrix = df_numerized.corr()

corr_pairs = correlation_matrix.unstack()

corr_pairs
```

```python
# Sort by absolute correlation value (most related pairs first)
sorted_pairs = corr_pairs.sort_values(ascending=False)

# Show the strongest correlations (excluding self-correlations of 1.0)
strong_correlations = sorted_pairs[sorted_pairs < 0.9999]
print("Top 20 Strongest Correlations:")
print(strong_correlations.head(20))
```

```python
# Positive correlations
positive_corr = corr_pairs[corr_pairs > 0].sort_values(ascending=False)
positive_corr = positive_corr[positive_corr < 1]  # Remove self-correlations

print("Top Positive Correlations:")
print(positive_corr.head(15))

# Negative correlations
negative_corr = corr_pairs[corr_pairs < 0].sort_values()

print("\nTop Negative Correlations:")
print(negative_corr.head(15))# Positive correlations
positive_corr = corr_pairs[corr_pairs > 0].sort_values(ascending=False)
positive_corr = positive_corr[positive_corr < 1]  # Remove self-correlations

print("Top Positive Correlations:")
print(positive_corr.head(15))

# Negative correlations
negative_corr = corr_pairs[corr_pairs < 0].sort_values()

print("\nTop Negative Correlations:")
print(negative_corr.head(15))
```

```python
# Reset index to get pairs as columns
corr_df = corr_pairs.reset_index()
corr_df.columns = ['Variable_1', 'Variable_2', 'Correlation']

# Remove self-correlations and duplicates (A-B is same as B-A)
corr_df = corr_df[corr_df['Variable_1'] != corr_df['Variable_2']]
corr_df['Pair'] = corr_df.apply(lambda x: tuple(sorted([x['Variable_1'], x['Variable_2']])), axis=1)
corr_df = corr_df.drop_duplicates('Pair').drop('Pair', axis=1)

# Sort by absolute correlation
corr_df['Abs_Correlation'] = abs(corr_df['Correlation'])
corr_df = corr_df.sort_values('Abs_Correlation', ascending=False)

print("Correlation Pairs DataFrame:")
corr_df.head(20)
```
