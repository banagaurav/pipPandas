```python
# creating a web scrapper using 
# checking the price of a book in books.toscrape.com 
# creating a dataset BookWebScrapperDataSet.csv and automating for everyday
#import libraries

from bs4 import BeautifulSoup
import requests
import time
import datetime
# import smtplibW
print("All imports working!")
```

```python
#connect to website

URL = "https://books.toscrape.com/catalogue/a-light-in-the-attic_1000/index.html"

hearders = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"}

page = requests.get(URL, headers = hearders)

soup1 = BeautifulSoup(page.content, "html.parser")

soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

title = soup2.find("div", class_="product_main").find("h1").get_text(strip=True)

price = soup2.find("p", class_="price_color").get_text(strip=True)[1:]

print(title)
print(price)
```

```python
def check_price():
    URL = "https://books.toscrape.com/catalogue/a-light-in-the-attic_1000/index.html"

    hearders = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"}

    page = requests.get(URL, headers = hearders)

    soup1 = BeautifulSoup(page.content, "html.parser")

    soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

    title = soup2.find("div", class_="product_main").find("h1").get_text(strip=True)

    price = soup2.find("p", class_="price_color").get_text(strip=True)[1:]

    import datetime

    today = datetime.date.today()
    print(today)

    import csv
    header = ['Title','Price','Date']
    data = [title, price, today]

    with open('BookWebScrapperDataSet.csv','a+',newline='',encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data) 
```
