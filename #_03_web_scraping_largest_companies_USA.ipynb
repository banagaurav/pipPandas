```python
from bs4 import BeautifulSoup
import requests
import pandas as pd 
```

```python
URL = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'


headers = {
    "User-Agent": "Mozilla/5.0 (learning-scraping; contact: your-email@example.com)"
}

page = requests.get(URL, headers = headers)

soup = BeautifulSoup(page.text, 'html')

print(soup)
```

```python
# this returns list so we can use indexing
# List of the largest public / publicly traded companies
table1 = soup.find_all('table')[0]
```

```python
# List of largest private companies
table2 = soup.find_all('table')[1]
table2
```

```python
# we could also find that table like this 

# table = soup.find('table', class_ = 'wikitable sortable')
# table
```

```python
print(table1)
```

```python
world_titles = table1.find_all('th')
```

```python
# in this strip is to eliminate anything else such as backspaces and take only text
world_table_title = [title.text.strip() for title in world_titles]
```

```python
# adding world_table_title as columns header in df 

df = pd.DataFrame(columns = world_table_title)

df
```

```python
column_data = table1.find_all('tr')
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row = [data.text.strip() for data in row_data]
    
    length_df = len(df)
    df.loc[length_df] = individual_row
```

```python
# export csv 
df.to_csv(r'D:\Code\PipPandas\outputs\public_companies.csv', index=False)
```

```python
# taking out table th
headers2 = table2.find_all('th')
headers_private_companies = [headers.text.strip() for headers in headers2]
headers_private_companies
```

```python
# adding headers_private_companies as columns header in df
df2 = pd.DataFrame(columns= headers_private_companies)
df2
```

```python
column_data2  = table2.find_all('tr')
for row in column_data2[1:]:
    row_data2 = row.find_all('td')
    individual_row2 = [data.text.strip() for data in row_data2]
    
    length_df2 = len(df2)
    df2.loc[length_df2] = individual_row2

df2

```

```python
df2.to_csv(r'D:\Code\PipPandas\outputs\private_companies.csv', index=False)
```
